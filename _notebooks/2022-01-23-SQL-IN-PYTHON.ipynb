```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pyodbc

conn_str = (
    r'DRIVER={SQL Server};'
    r'SERVER=SEIFEDIN\SQLEXPRESS;'
    r'DATABASE=Fifa 18;'
    r'Trusted_Connection=yes;'
)
cnxn = pyodbc.connect(conn_str)
query = "SELECT * FROM Fifa18"
cursor = cnxn.cursor()
# cursor.execute("SELECT * FROM Fifa18")

df = pd.read_sql(query, cnxn)
df
```

```python
df.head(5)
```

### Let's check how many countries are represented in the dataset.



```python


Nationality = "SELECT COUNT(DISTINCT Nationality) AS Nationality FROM Fifa18"

df= pd.read_sql(Nationality, cnxn)
df
```

### Let's chech how many players are in the FIFA 

```python
Num_players = "SELECT COUNT(*) AS Num_Of_Players FROM Fifa18"

df= pd.read_sql(Num_players, cnxn)

df
```

#### There are 17981 players in the FIFA

### Find out how many players are from each country...

```python
NumOfPlayerPerCuntry = "SELECT COUNT(*) AS NumOfPlayerPerCuntry, Nationality FROM Fifa18 GROUP BY Nationality ORDER BY NumOfPlayerPerCuntry DESC"
                       

df= pd.read_sql(NumOfPlayerPerCuntry, cnxn)

df.head(10)
```

### What's the highest Salary paid to a player? 


```python
Highest_Salary = "SELECT MAX([Wage (€)]) AS Max_Salary FROM Fifa18"
df= pd.read_sql(Highest_Salary, cnxn)
df
```

### Which player is getting paid the highest and where is he from?
    

```python
HighestPaidPlayer = "SELECT Name, Nationality FROM Fifa18 WHERE [Wage (€)] = 565000"
df= pd.read_sql(HighestPaidPlayer, cnxn)
df
```

#### It appears we have ties. The two GOATS(CR7 and L.Messi Make the same amount)

```python
# Let's using SQL neesting to perform the above query.
HighestPaidPlayer = "SELECT Name, [Wage (€)], Nationality  FROM Fifa18 WHERE [Wage (€)] = (SELECT MAX([Wage (€)]) FROM Fifa18)"
df= pd.read_sql(HighestPaidPlayer, cnxn)
df
```

#### What's the minimum Salary paid?

```python
Min_Salary = "SELECT Min([Wage (€)]) FROM Fifa18"
df= pd.read_sql(Min_Salary, cnxn)
df
```

It appears the are some people who are just volunteers.

### Which player is got the highest rating and from which club? USE THE NESTED QUERY METHOD FOR THIS.

```python
Top_player = "SELECT Name, Overall, Club FROM Fifa18 WHERE Overall = (SELECT MAX(Overall) FROM Fifa18)"
df = pd.read_sql(Top_player, cnxn)
df
```

#### Cristiano Ronaldo is the highest rated player in FIFA

### Which are top 3 clubs based on overall rating ?

```python
Top_Clubs = "SELECT TOP 4 AVG(Overall) AS Avg_Overall_Rating, Club FROM Fifa18  GROUP BY Club ORDER BY AVG(Overall) DESC"
df = pd.read_sql(Top_Clubs, cnxn)
df
```

### We can even combine with Viz from here. We can run both Matplotlib or Seaborn right here at the same time

```python
fig = plt.figure()
ax = fig.add_axes([5,5,1,1])
x_ax = df["Club"]
y_ax = df["Avg_Overall_Rating"]
ax.bar(x_ax, y_ax)

ax.set_xlabel("Clubs")
ax.set_ylabel("Overall Rating")
ax.set_title("Highest Rated Clubs in Fifa")
plt.show
```

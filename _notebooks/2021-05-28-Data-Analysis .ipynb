# Aggregating the rows by country: 
**Instead of having multiple data from the same country based on the province/region, we can combine them and get a signle dataset for each country.**

```python
corona_dataset_aggregated= corona_dataset_csv.groupby('Country/Region').sum()  # This will group data from each region and sum up the total to result in a single output for each country
corona_dataset_aggregated
```

```python
corona_dataset_aggregated.loc['Spain'].diff().max()
```

```python
#Let's calculate the maximum infection rate for each country and assign it to its own col.
#Since the indexes of our dataset is a list of coutries,we can do as follow: 

countries = list(corona_dataset_aggregated.index)

#Create an empty list and append the result of each countries infection rate into the new list 

max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
  
max_infection_rates
    
    
```

```python
#The above result is maximum infection rates for each country.Let's put it into its own column. 
```

```python
corona_dataset_aggregated['max_infection_rate'] = max_infection_rates
```
